```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('D:\GitHub Repos\culminating-project-group-1\Week 2\Database Files\BANA698GROUP1.db.db')
df = pd.read_sql_query('SELECT "CountryShortName", Year, "IndicatorName", Value FROM AllRecords', conn)

pivot_df = df.pivot_table(index=["CountryShortName", "Year"], columns="IndicatorName", values="Value", aggfunc='max')
print(pivot_df)

pivot_df.to_csv('Group1Data.csv')
```

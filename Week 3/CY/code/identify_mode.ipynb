```python
#this identifies features that have an unusual number of unique values compared to the norm
#if data is supposed to be uniform such that, for example, there's 20 rows per country (one per year)
#yet one country is short a few years, this can spot it
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
import gc
def del_df(*df_names):
    for df_name in df_names:
        if df_name in globals():
            del globals()[df_name]
    
    gc.collect()
```

```python
folder_loc = 'F:/class/BANA 698/week 4/rawdata is 20 years/with2024/'
#gets a list of the files in the folder; from the output, copy the region and income data sets only for making a list below
import os
file_list = [f for f in os.listdir(folder_loc) if os.path.isfile(os.path.join(folder_loc, f))]
for file in file_list:
    print(file)
```

```python
file_list = [
    'Group1Data_20years.csv',
    'Group1Data_20years_only_eastasiapacific.csv',
    'Group1Data_20years_only_europecentralasia.csv',
    'Group1Data_20years_only_highincome.csv',
    'Group1Data_20years_only_latinamericacaribbean.csv',
    'Group1Data_20years_only_lowermiddleincome.csv',
    'Group1Data_20years_only_lowincome.csv',
    'Group1Data_20years_only_middleeastnorthafrica.csv',
    'Group1Data_20years_only_northamerica.csv',
    'Group1Data_20years_only_southasia.csv',
    'Group1Data_20years_only_subsaharanafrica.csv',
    'Group1Data_20years_only_uppermiddleincome.csv'
]
```

```python
#identify unique values that don't occur as much as the other values
#eg if most countries have 20 years' worth of data, then the mode is 20. any country that doesn't show up 20 times (once per year) is flagged.
#eg if most countries have the same years of data, then each numerical Year is the mode. any year that doesn't occur as often is flagged.

column = 'Year'
#column = 'CountryShortName'

for file in file_list:
    df = pd.read_csv(folder_loc + file)
    value_counts = df[column].value_counts()
    mode_count = value_counts.mode()[0]
    unusual_counts = value_counts[value_counts != mode_count]
 
    print(f"\nUnusual counts for {column} in {file} (mode is {mode_count})")
    print(unusual_counts)
    del_df('df')
```

```python
#display count of uniques for a column

column = 'Year'
#column = 'CountryShortName'

for file in file_list:
    df = pd.read_csv(folder_loc + file)
    value_counts = df[column].value_counts()

    print(f"\nValue counts for {column} in {file}")
    print(value_counts)
    del_df('df')
```
